##### Exercise Set 3.1

8. There is a real number x such that x > 1 and 2^x > x^10.  

```python
2**1.01 > 1.01**10
```

Finding an single answer was easy but I want to find the point where 2^x == x^10.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1.25, 100)

fig, ax = plt.subplots()
ax.plot(x, 2**x, label="2^x")
ax.plot(x, x**10, label="x^10")
ax.set_xlabel("input")
ax.set_ylabel("result")
ax.legend()
```
